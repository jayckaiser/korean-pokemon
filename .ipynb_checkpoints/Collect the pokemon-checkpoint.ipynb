```python
import csv
import os
import re
import requests
import time
import urllib.request
```

```python
page_url = "https://www.serebii.net/pokedex-xy/{}.shtml"
image_url = "https://www.serebii.net/xy/pokemon/{}.png"

image_filepath = "./images/{}.png"
output_filepath = "pokemon_names.csv"
```

```python
pokemon_dict = {}
```

```python
def do_the_thing(num):
    z_num = str(num).zfill(3)
    
    try:
        # Save the image (if not saved already.)
        if not os.path.exists(image_filepath):
            urllib.request.urlretrieve(image_url.format(z_num),
                                       image_filepath.format(z_num))

        page = requests.get(page_url.format(z_num))
        raw_text = page.text

        # Retrieve the Korean name
        raw_korean = re.findall(r'Korean</b>: </td><td>(.*)</td>', raw_text)[0]
        raw_korean = raw_korean.replace(";", "")
        real_korean = ''.join(chr(int(u)) for u in raw_korean.split('&#') if u)

        # Retrieve the English name
        real_english = re.findall(r'<title>(.*) - #', raw_text)[0]

        output_tuple = (real_korean, real_english)
        pokemon_dict[z_num] = output_tuple
        
        print("{} -- {} -- {}".format(z_num, *output_tuple))
    
    except Exception as err:
        print("Failed on {} :: {}".format(z_num, err))
```

```python
for num in range(1, 722):
    
    do_the_thing(num)
    
    time.sleep(1)
```

```python
do_the_thing(721)
```

```python
# Make some manual repairs.
pokemon_dict['029'] = ('니드런♀', 'Nidoran♀')
pokemon_dict['032'] = ('니드런♂', 'Nidoran♂')
pokemon_dict['233'] = ('폴리곤2', 'Porygon2')
pokemon_dict['474'] = ('폴리곤Z', 'Porygon-Z')
```

```python
pokemon_dict
```

```python
# Finally, save the list as a CSV file.
with open(output_filepath, 'w') as FILE:
    csv_out = csv.writer(FILE)
    csv_out.writerow(['index', 'korean', 'english'])
    
    sorted_dict = sorted(pokemon_dict.items(), key=lambda x: x[0])
    for k, v in sorted_dict:
        
        row = (k, v[0], v[1])
        csv_out.writerow(row)
```
